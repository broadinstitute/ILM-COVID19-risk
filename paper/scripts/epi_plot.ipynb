```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1f = pd.read_csv('set1_full/infected_number.txt')
data1f[data1f["cycle"] < 400][["susceptible", "latent", "infected", "recovered"]].plot()
print(max(data1f["latent"]), max(data1f["infected"]))
```

```python
import matplotlib.ticker as tkr 

def days_fmt(x, pos): # your custom formatter function: divide by 4
    s = '{}'.format(int(x / 4))
    return s

dfmt = tkr.FuncFormatter(days_fmt)
```

```python
var = "latent"
tlim = 300

dat = data1f[data1f["cycle"] < tlim]
l1f = np.array(dat[var])

dat = data1q_delayed[data1q_delayed["cycle"] < tlim]
l1qd = np.array(dat[var])

dat = data1q_nodelay[data1q_nodelay["cycle"] < tlim]
l1qn = np.array(dat[var])

x = np.arange(len(l1f))

fig, ax = plt.subplots(figsize=(12,5))
ax.xaxis.set_major_formatter(dfmt)
plt.plot(x, l1f, linewidth=2, label='No quarantine')
plt.plot(x, l1qd, linewidth=2, label='Delayed quarantine')
plt.plot(x, l1qn, linewidth=2, label='Instant quarantine')
plt.xlabel('Days')
plt.ylabel('Cases')
plt.legend()
plt.tight_layout()
fig.savefig('risk_quarantine1.pdf')
```
